```python
import operator
import tree ## use already defined tree class
## In the tree class, using Node class to represent each node. All the nodes form the tree. 
## Node.label returns a string, which is the value in the node.
## Node.children returns a list of Nodes, which are the children of the current node.
```

```python
rules = [] ## Store all the rules in the training file, each rule is represented as a string
with open("C:/Users/Xin/Desktop/Github-repo/Natural_Language_Processing_Projects/Project_4_Build_Parser/hw4-data/train.trees.pre.unk") as train_trees:
    for line in train_trees:
        t = tree.Tree.from_str(line) ## Transform the string to the Tree data structure
        for node in t.bottomup(): ## Interate all the node in each tree
            if node.children: ## exclue the terminal node
                rule = node.label
                for child in node.children:
                    rule = rule + " " + child.label
                rules.append(rule)        
```

```python
## Count the frequency of the rules
rule_count={}
for i in rules:
    if i not in rule_count:
        rule_count[i] = 1
    else:
        rule_count[i] += 1
        
print(len(rule_count))
```

```python
## Sort the rule_count dictionary by values
rule_count_sorted = sorted(rule_count.items(), key=operator.itemgetter(1),reverse=True)
```

```python
## Print out the top five most frequent rules
for i in range(5):
    print(rule_count_sorted[i])
```

```python
## Calculate the conditional probability of each rule
count_nonterminal_dict = {}
for i in rule_count:
    i_list = i.split()
    if i_list[0] not in count_nonterminal_dict:
        count_nonterminal_dict[i_list[0]] = rule_count[i]
    else:
        count_nonterminal_dict[i_list[0]] += rule_count[i]
rule_frequency = {}
for i in rule_count:
    i_list = i.split()
    rule_frequency[i] = rule_count[i]/count_nonterminal_dict[i_list[0]]
```

```python
## Sort the rule_frequency dictionary by values
rule_frequency_sorted = sorted(rule_frequency.items(), key=operator.itemgetter(1),reverse=True)
```

```python
## Print out the top five most frequent rules
for i in range(40):
    print(rule_frequency_sorted[i])
```

```python
## Print out the rule probability according to the given format
for i in rule_frequency_sorted:
    i_list = i[0].split()
    if len(i_list) == 2:
        output = i_list[0]+" "+"-"+">"+" "+i_list[1]+" #"+" "+str(i[-1])
    if len(i_list) == 3:
        output = i_list[0]+" "+"-"+">"+" "+i_list[1]+" "+i_list[2] + " #" + " " + str(i[-1])
    print(output)
    
```
